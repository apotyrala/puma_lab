```
#2a cd-correlation
import numpy as np, pandas as pd; 
import seaborn as sns;
import io
import requests

#get the data and convert it to pandas dataframe (split in rows and columns)

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
s = requests.get(url).content
s = s.decode('utf-8')
s_rows = s.split('\n')
s_rows_cols = [each.split(",") for each in s_rows]
c = pd.DataFrame(s_rows_cols, columns = names)

fa,sa=getparams()

correlation = c[c.columns[fa]].astype('float64').corr(c[c.columns[sa]].astype('float64'))
print(correlation)
```

```
#2a cd-covariance

fa,sa=getparams()

covariance= c[c.columns[fa]].astype('float64').cov(c[c.columns[sa]].astype('float64'))
print(covariance)
```
